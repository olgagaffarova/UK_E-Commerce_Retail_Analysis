# Q1: How much should we scale bikes back between November and April?

## Approach: Fleet Scaling (Nov–Apr)

**Goal:**  
Determine how much to reduce the active fleet in low season (Nov–Apr) based on monthly demand.

**Steps:**

1. **Aggregate monthly trips** for each bike type.
2. **Find the peak month** (100% demand baseline) for each type.
3. **Calculate demand_vs_peak_%:**  
   monthly_rides / peak_rides × 100.
4. **Compute fleet_reduction_%:**  
   100 − demand_vs_peak_% → how much demand drops that month.
5. **Select low->season months (Nov–Apr)** and compute average reduction:
   - Classic: ~58%  
   - Electric: ~44%
6. **Convert to system-wide recommendation:**  
   weighted reduction ≈ ~32% → operational range **30–40%**.

**Result:**  
A simple, justified guideline: reduce active fleet by **30–40%** during Nov–Apr based on actual seasonal demand patterns.

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from datetime import datetime as dt
import os
import seaborn as sns
```

```python
# Loading data
essential_cols = ['month', 'rideable_type', 'ride_id']

df = pd.read_pickle(r'../02_Data/df_merged_3.0.pkl')[essential_cols]
```

```python
# Step 1: Aggregate total rides by month
monthly_rides = (
    df.groupby('month', as_index=False)['ride_id']
    .count()
    .rename(columns={'ride_id': 'total_rides'})
)

# Optional: sort months chronologically 
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
```

```python
# ============================================================================
# MONTHLY FLEET SCALING ANALYSIS — BY BIKE TYPE
# ============================================================================

# Step 1: Aggregate total rides by month and bike type
monthly_type = (
    df.groupby(['month', 'rideable_type'], as_index=False)['ride_id']
    .count()
    .rename(columns={'ride_id': 'total_rides'})
)
monthly_type['month'] = pd.Categorical(monthly_type['month'], categories=month_order, ordered=True)
monthly_type = monthly_type.sort_values(['rideable_type', 'month'])

# Step 2: Calculate peak rides for each bike type
monthly_type['peak_rides'] = monthly_type.groupby('rideable_type')['total_rides'].transform('max')

# Step 3: Calculate demand as percentage of peak
monthly_type['demand_vs_peak_%'] = (monthly_type['total_rides'] / monthly_type['peak_rides']) * 100

# Step 4: Calculate recommended fleet reduction
# Logic: If demand is only 40% of peak, you can reduce fleet by 60% (100 - 40)
monthly_type['fleet_reduction_%'] = 100 - monthly_type['demand_vs_peak_%']

# Round for readability
monthly_type['demand_vs_peak_%'] = monthly_type['demand_vs_peak_%'].round(1)
monthly_type['fleet_reduction_%'] = monthly_type['fleet_reduction_%'].round(1)

# Step 5: Filter low-season months (Nov–Apr)
low_season = ['November', 'December', 'January', 'February', 'March', 'April']
low_season_type = monthly_type[monthly_type['month'].isin(low_season)].copy()

# Step 6: Create summary by bike type for low season
low_season_summary = (
    low_season_type.groupby('rideable_type')
    .agg({
        'fleet_reduction_%': 'mean',  # Average reduction across low season
        'total_rides': 'sum'
    })
    .round(1)
    .rename(columns={
        'fleet_reduction_%': 'avg_fleet_reduction_%',
        'total_rides': 'total_low_season_rides'
    })
)

# Display results
print("=" * 80)
print("📊 MONTHLY FLEET REDUCTION RECOMMENDATIONS BY BIKE TYPE")
print("=" * 80)
print("\nFull Year Analysis:")
print(monthly_type[['month', 'rideable_type', 'total_rides', 'demand_vs_peak_%', 'fleet_reduction_%']].to_string(index=False))

print("\n" + "=" * 80)
print("🧊 LOW-SEASON (Nov–Apr) DETAILED BREAKDOWN")
print("=" * 80)
print(low_season_type[['month', 'rideable_type', 'total_rides', 'demand_vs_peak_%', 'fleet_reduction_%']].to_string(index=False))

print("\n" + "=" * 80)
print("📈 LOW-SEASON SUMMARY BY BIKE TYPE")
print("=" * 80)
print(low_season_summary)
```

```python
low_season_summary.to_csv("low_season_summary.csv")
```

```python
low_season_type.to_csv("low_season_type.csv")
```

```python
monthly_type.to_csv("fleet_reduction.csv")
```

```python
# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 12)

# ============================================================================
# CREATE VISUALIZATIONS FOR FLEET REDUCTION DASHBOARD
# ============================================================================

# Create figure with subplots
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

# Color palette for bike types
bike_colors = {
    'electric_bike': '#FF6B6B',
    'classic_bike': '#4ECDC4',
}

# Get unique bike types
bike_types = monthly_type['rideable_type'].unique()

# ============================================================================
# 1. LINE CHART: Monthly Demand vs Peak % by Bike Type
# ============================================================================
ax1 = fig.add_subplot(gs[0, :])
for bike_type in bike_types:
    data = monthly_type[monthly_type['rideable_type'] == bike_type]
    ax1.plot(data['month'], data['demand_vs_peak_%'], 
             marker='o', linewidth=2.5, markersize=8,
             label=bike_type.replace('_', ' ').title(),
             color=bike_colors.get(bike_type, '#333'))

ax1.axhline(y=100, color='green', linestyle='--', alpha=0.5, label='Peak Capacity (100%)')
ax1.axhline(y=50, color='orange', linestyle='--', alpha=0.5, label='50% Threshold')
ax1.fill_between(range(len(month_order)), 0, 100, 
                  where=np.isin(month_order, low_season),
                  alpha=0.1, color='blue', label='Low Season')
ax1.set_xlabel('Month', fontsize=12, fontweight='bold')
ax1.set_ylabel('Demand (% of Peak)', fontsize=12, fontweight='bold')
ax1.set_title('Monthly Demand Patterns by Bike Type', fontsize=14, fontweight='bold', pad=20)
ax1.legend(loc='upper right', frameon=True)
ax1.grid(True, alpha=0.3)
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')

# ============================================================================
# 2. HEATMAP: Fleet Reduction Recommendations
# ============================================================================
ax2 = fig.add_subplot(gs[1, :])
pivot_data = monthly_type.pivot(index='rideable_type', columns='month', values='fleet_reduction_%')
pivot_data.index = pivot_data.index.str.replace('_', ' ').str.title()

sns.heatmap(pivot_data, annot=True, fmt='.0f', cmap='RdYlGn_r', 
            cbar_kws={'label': 'Fleet Reduction %'}, 
            linewidths=0.5, linecolor='white',
            vmin=0, vmax=100, ax=ax2)
ax2.set_title('Fleet Reduction Recommendations (%) — Heatmap View', 
              fontsize=14, fontweight='bold', pad=20)
ax2.set_xlabel('Month', fontsize=12, fontweight='bold')
ax2.set_ylabel('Bike Type', fontsize=12, fontweight='bold')

# ============================================================================
# 3. BAR CHART: Average Low-Season Reduction by Type
# ============================================================================
ax3 = fig.add_subplot(gs[2, 0])
low_season_avg = low_season_type.groupby('rideable_type')['fleet_reduction_%'].mean().sort_values(ascending=True)
colors = [bike_colors.get(bt, '#333') for bt in low_season_avg.index]
bars = ax3.barh(low_season_avg.index.str.replace('_', ' ').str.title(), 
                low_season_avg.values, color=colors, edgecolor='black', linewidth=1.5)

# Add value labels
for i, (bar, val) in enumerate(zip(bars, low_season_avg.values)):
    ax3.text(val + 1, i, f'{val:.1f}%', va='center', fontweight='bold')

ax3.set_xlabel('Average Fleet Reduction (%)', fontsize=11, fontweight='bold')
ax3.set_title('Low-Season Average Reduction\n(Nov–Apr)', fontsize=12, fontweight='bold')
ax3.grid(axis='x', alpha=0.3)

# ============================================================================
# 4. STACKED AREA: Total Rides by Type Over Time
# ============================================================================
ax4 = fig.add_subplot(gs[2, 1])
pivot_rides = monthly_type.pivot(index='month', columns='rideable_type', values='total_rides').fillna(0)
pivot_rides.plot.area(ax=ax4, alpha=0.7, 
                      color=[bike_colors.get(col, '#333') for col in pivot_rides.columns],
                      linewidth=2)

ax4.set_xlabel('Month', fontsize=11, fontweight='bold')
ax4.set_ylabel('Total Rides', fontsize=11, fontweight='bold')
ax4.set_title('Ride Volume Distribution by Type', fontsize=12, fontweight='bold')
ax4.legend(title='Bike Type', labels=[col.replace('_', ' ').title() for col in pivot_rides.columns])
plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax4.grid(True, alpha=0.3)

# Format y-axis with comma separator
ax4.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))

# Add overall title
fig.suptitle('🚴 FLEET SCALING DASHBOARD — Q1 Analysis: Low-Season Reduction Strategy', 
             fontsize=16, fontweight='bold', y=0.995)

plt.tight_layout()
plt.show()

# ============================================================================
# summary statistics box
# ============================================================================
print("\n" + "="*80)
print("📊 KEY INSIGHTS FOR DASHBOARD")
print("="*80)

for bike_type in bike_types:
    bike_data = low_season_type[low_season_type['rideable_type'] == bike_type]
    avg_reduction = bike_data['fleet_reduction_%'].mean()
    min_reduction = bike_data['fleet_reduction_%'].min()
    max_reduction = bike_data['fleet_reduction_%'].max()
    
    print(f"\n{bike_type.replace('_', ' ').title()}:")
    print(f"  • Average Low-Season Reduction: {avg_reduction:.1f}%")
    print(f"  • Range: {min_reduction:.1f}% - {max_reduction:.1f}%")
    print(f"  • Least impacted month: {bike_data.loc[bike_data['fleet_reduction_%'].idxmin(), 'month']}")
    print(f"  • Most impacted month: {bike_data.loc[bike_data['fleet_reduction_%'].idxmax(), 'month']}")
```
