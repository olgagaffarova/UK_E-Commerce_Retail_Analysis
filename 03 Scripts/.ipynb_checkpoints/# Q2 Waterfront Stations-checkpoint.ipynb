# Q2 Waterfront Stations

```python
# Q2: How to determine how many more stations to add along the water?
# ============================================================================
# APPROACH OVERVIEW:
# This analysis uses a supply-demand framework to identify capacity gaps.
# 
# Key Methodology:
# 1. Define "waterfront" stations based on longitude thresholds
#    (Hudson River: < -74.01, East River: > -73.95)
# 2. Calculate supply: % of stations near water
# 3. Calculate demand: % of trips involving waterfront stations
# 4. Gap analysis: If demand > supply, waterfront is underserved
# 5. Time concentration: Identify peak-hour pressure points
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from datetime import datetime as dt
from streamlit_keplergl import keplergl_static
import os
import geopandas as gpd
```

```python
# Loading data
essential_cols = ['start_station_name', 'end_station_name',
    'start_lat', 'start_lng', 'end_lat', 'end_lng',
    'member_casual', 'rideable_type', 'month', 'day_type',
    'value', 'ride_duration']

df = pd.read_pickle(r'../02_Data/df_merged_3.0.pkl')[essential_cols]
```

```python
# ----------------------------------------------------------------------------
# BUILD UNIQUE STATION LIST (this creates `stations`)
# ----------------------------------------------------------------------------
stations = (
    df[['start_station_name', 'start_lat', 'start_lng']]
    .drop_duplicates(subset='start_station_name')  # remove duplicates
    .rename(columns={
        'start_station_name': 'station_name',
        'start_lat': 'lat',
        'start_lng': 'lng'
    })
    .reset_index(drop=True)
)

print("✅ stations dataframe created")
print(stations.head())
print("Total unique stations:", len(stations))
```

```python
# ----------------------------------------------------------------------------
# STEP 1: IDENTIFY WATERFRONT STATIONS (SUPPLY)
# ----------------------------------------------------------------------------
# Logic:
# - A station is considered "waterfront" if it is located within 300 meters
#   of either the Hudson River (west) or the East River (east).
# ----------------------------------------------------------------------------


# ----------------------------------------------------------------------------
# Convert to GeoDataFrame (needed for spatial calculations)
# ----------------------------------------------------------------------------
gdf_st = gpd.GeoDataFrame(
    stations,
    geometry=gpd.points_from_xy(stations['lng'], stations['lat']),
    crs="EPSG:4326"
)

# ----------------------------------------------------------------------------
# Load shoreline (NYC_Shoreline.geojson downloaded from NYC Open Data)
# ----------------------------------------------------------------------------
coastline = gpd.read_file("../02_Data/NYC_Shoreline.geojson")

# ----------------------------------------------------------------------------
# Reproject both to meters (required for distance calculations)
# ----------------------------------------------------------------------------
gdf_st = gdf_st.to_crs(epsg=3857)
coastline = coastline.to_crs(epsg=3857)


# ----------------------------------------------------------------------------
# Classify waterfront stations based on distance to shoreline
# ----------------------------------------------------------------------------
DIST_THRESHOLD = 400  

gdf_st["dist_to_water_m"] = gdf_st.geometry.apply(
    lambda geom: coastline.distance(geom).min()
)

gdf_st["is_waterfront"] = gdf_st["dist_to_water_m"] <= DIST_THRESHOLD


# ----------------------------------------------------------------------------
#  Supply summary
# ----------------------------------------------------------------------------
total_stations = len(gdf_st)
waterfront_station_count = int(gdf_st["is_waterfront"].sum())
supply_share = waterfront_station_count / total_stations

print("\n--- SUPPLY ANALYSIS (shoreline-based) ---")
print(f"Total stations: {total_stations:,}")
print(f"Waterfront stations (≤ {DIST_THRESHOLD}m to shoreline): {waterfront_station_count:,}")
print(f"Waterfront supply share: {supply_share:.1%}")
```

```python
# ----------------------------------------------------------------------------
# VISUALIZE WATERFRONT STATIONS ON MAP (Plotly Mapbox)
# ----------------------------------------------------------------------------
import plotly.express as px

# Convert back to WGS84 lat/lon for mapping (EPSG:4326)
gdf_plot = gdf_st.to_crs(epsg=4326)

fig = px.scatter_mapbox(
    gdf_plot,
    lat="lat",
    lon="lng",
    color="is_waterfront",
    hover_name="station_name",
    hover_data={"dist_to_water_m": True, "lat": False, "lng": False},
    color_discrete_map={True: "#1f77b4", False: "#ff7f0e"},  # blue = waterfront, orange = non
    zoom=10.8,
    height=700,
    title=f"Waterfront vs. Non-Waterfront CitiBike Stations (≤ {DIST_THRESHOLD}m to shoreline)"
)

fig.update_layout(
    mapbox_style="open-street-map",  # no token required
    legend=dict(
        title="Station Type",
        itemsizing="constant"
    ),
    margin=dict(l=0, r=0, t=60, b=0)
)

fig.show()
```

```python
# ============================================================================
# KEPLER MAP — Waterfront vs Non-Waterfront CitiBike Stations
# ============================================================================

from keplergl import KeplerGl
import geopandas as gpd

# Always export data to EPSG:4326 (lat/lon)
gdf_kepler = gdf_st.to_crs(epsg=4326).copy()

# Kepler expects plain columns lat/lng, not geometry only
gdf_kepler["latitude"] = gdf_kepler.geometry.y
gdf_kepler["longitude"] = gdf_kepler.geometry.x

# Build Kepler map
m = KeplerGl(height=750)

m.add_data(
    data=gdf_kepler[["station_name", "latitude", "longitude", "is_waterfront", "dist_to_water_m"]],
    name="CitiBike Stations"
)

# OPTIONAL — automatically center map over NYC
m.config = {
  "mapState": {
    "bearing": 0,
    "latitude": gdf_kepler["latitude"].mean(),
    "longitude": gdf_kepler["longitude"].mean(),
    "pitch": 0,
    "zoom": 10.5
  }
}

m
```

```python
# ============================================================================
# STEP 2: CALCULATE DEMAND (Usage)
# ============================================================================
# Demand = % of all trips that start OR end at a waterfront station
# ============================================================================

# List of waterfront station names (based on shoreline distance classification)
waterfront_list = set(gdf_st.loc[gdf_st['is_waterfront'], 'station_name'])

# Create boolean flags for each trip
df['start_is_waterfront'] = df['start_station_name'].isin(waterfront_list)
df['end_is_waterfront']   = df['end_station_name'].isin(waterfront_list)

# A trip "touches" waterfront if start OR end is at a waterfront station
df['trip_touches_water'] = df['start_is_waterfront'] | df['end_is_waterfront']

# ----------------------------
# OVERALL DEMAND
# ----------------------------
waterfront_trip_count = int(df['trip_touches_water'].sum())
demand_share = df['trip_touches_water'].mean()

print("\n--- DEMAND ANALYSIS (shoreline-based) ---")
print(f"Trips touching waterfront: {waterfront_trip_count:,}")
print(f"Waterfront demand share: {demand_share:.1%}")


# ----------------------------
# MONTHLY DEMAND (% of trips touching water per month)
# ----------------------------
monthly_demand = (
    df.groupby("month")["trip_touches_water"]
      .mean()
      .reset_index()
      .rename(columns={"trip_touches_water": "monthly_demand_share"})
)

print("\n--- MONTHLY WATERFRONT DEMAND SHARE ---")
print(monthly_demand)
```

```python
monthly_demand["supply_share"] = supply_share
```

```python
monthly_demand["gap"] = monthly_demand["monthly_demand_share"] - monthly_demand["supply_share"]
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Line for demand (varies month to month)
fig.add_trace(go.Scatter(
    x=monthly_demand["month"],
    y=monthly_demand["monthly_demand_share"],
    mode="lines+markers",
    name="Monthly Demand (% trips touching waterfront)",
    line=dict(width=3, color="#ff7f0e"),
    marker=dict(size=8)
))

# Flat line for supply (constant)
fig.add_trace(go.Scatter(
    x=monthly_demand["month"],
    y=monthly_demand["supply_share"],
    mode="lines",
    name="Waterfront Supply (% of stations on waterfront)",
    line=dict(width=4, dash="dash", color="#1f77b4")   # <-- thicker + dashed
))

fig.update_layout(
    title="Waterfront Demand vs Supply — By Month",
    yaxis_tickformat=".0%",
    xaxis_title="Month",
    yaxis_title="% of network",
    height=500,
    legend=dict(title="Metrics"),
)

# ✅ Force the y-axis to show supply clearly
fig.update_yaxes(range=[0.10, 0.30])  # <-- change as needed to fit your values

fig.show()
```

```python
# ============================================================================
# STEP 3: GAP ANALYSIS — Do we need more waterfront stations?
# ============================================================================
# gap > 0  → demand exceeds supply → waterfront underserved
# gap < 0  → supply exceeds demand → no new stations required
# ============================================================================

gap = demand_share - supply_share

# Only estimate missing stations if supply < demand
estimated_missing_stations = max(int(gap * total_stations), 0)

print("\n--- GAP ANALYSIS (shoreline-based) ---")
print(f"Supply share: {supply_share:.1%}")
print(f"Demand share: {demand_share:.1%}")
print(f"Gap (Demand - Supply): {gap:+.1%}")

if estimated_missing_stations > 0:
    print(f"➡ Recommendation: Add ~{estimated_missing_stations} new stations along the waterfront.")
else:
    print("✅ Current station count meets or exceeds demand — no new stations needed.")
```

```python
# ============================================================================
# STEP 4: HOTSPOTS — Which waterfront areas have the highest pressure?
# ============================================================================
# Identify OD (origin → destination) pairs that touch the shoreline
# ============================================================================
waterfront_trips = (
    df[df['trip_touches_water']]      # only trips that start/end at waterfront
    .groupby(['start_station_name', 'end_station_name'], as_index=False)['value']
    .sum()
    .sort_values(by='value', ascending=False)
)

print("\n✅ HOTSPOTS dataframe created: waterfront_trips")
print(waterfront_trips.head(10))
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Indicator(
    mode="number+delta",
    value=demand_share,
    delta={'reference': supply_share, 'relative': True, 'valueformat': ".1%"},
    number={'valueformat': '.1%'},
    title={"text": "Waterfront Demand vs Supply<br><span style='font-size:14px'>Gap (Demand - Supply)</span>"}
))

fig.update_layout(height=300)
fig.show()
```

```python
import plotly.graph_objects as go

# take top 20 busiest OD pairs
top_flows = waterfront_trips.head(20)

# build node labels
nodes = list(set(top_flows['start_station_name']).union(top_flows['end_station_name']))

# map station names to index positions
node_index = {station: idx for idx, station in enumerate(nodes)}

# create Sankey structure
sankey_data = dict(
    type='sankey',
    node=dict(
        pad=20,
        thickness=15,
        line=dict(color="black", width=0.5),
        label=nodes
        
    ),
    link=dict(
        source=[node_index[s] for s in top_flows["start_station_name"]],
        target=[node_index[e] for e in top_flows["end_station_name"]],
        value=top_flows["value"]
    )
)

fig = go.Figure(data=[sankey_data])

fig.update_layout(
    title="Top Waterfront Origin → Destination Flows (Stations with highest trip pressure)",
    height=800
)

fig.show()
```

```python
# ============================================================================
# ORIGIN–DESTINATION MATRIX (HEATMAP) — FIXED LABEL CUT-OFF
# ============================================================================

import plotly.express as px
import pandas as pd

# Create a pivot table: rows = start stations, columns = end stations
df_matrix = top_flows.pivot_table(
    index='start_station_name',
    columns='end_station_name',
    values='value',
    fill_value=0
)

# Optional: sort by total flow
df_matrix = df_matrix.loc[df_matrix.sum(axis=1).sort_values(ascending=False).index]

# Optional: shorten labels slightly (if very long)
df_matrix.index = [name[:25] + "..." if len(name) > 25 else name for name in df_matrix.index]
df_matrix.columns = [name[:25] + "..." if len(name) > 25 else name for name in df_matrix.columns]

# Create heatmap
fig = px.imshow(
    df_matrix,
    color_continuous_scale='Oranges',
    aspect='auto',
    labels=dict(x="Destination Station", y="Origin Station", color="Trips"),
    title="Origin–Destination Matrix — Top Waterfront Routes"
)

# --- Layout tweaks for readability ---
fig.update_layout(
    width=1000,                      # make much wider
    height=600,                     # taller to fit y labels
    margin=dict(l=200, r=200, t=80, b=200),  # more padding for long labels
    xaxis=dict(
        showticklabels=True,
        tickangle=90,                # vertical labels
        tickfont=dict(size=11, color="black"),
        automargin=True
    ),
    yaxis=dict(
        showticklabels=True,
        tickfont=dict(size=11, color="black"),
        automargin=True
    ),
    title=dict(font=dict(size=22, color="black")),
    plot_bgcolor='white',
    paper_bgcolor='white'
)

fig.show()
```
