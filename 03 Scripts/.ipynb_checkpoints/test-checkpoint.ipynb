### Analysis Goal: Identifying Problem Stations for Bike Shortage Reduction

This part of the analysis focuses on finding **the top problem docks** — stations that consistently *run out of bikes* during the day.

We use two criteria:

1. **Top 20% busiest stations** (634 high-demand stations)  
   These stations collectively represent **80% of all CitiBike trips** (Pareto principle).

2. **Net_flow imbalance (starts − ends)**  
   - Positive net_flow → more trips **start** than end → bikes leave → **shortage risk**
   - Negative net_flow → more trips **end** than start → bikes accumulate → **overflow**

By calculating the **average daily net_flow** per station, we can detect which stations:

- consistently *lose bikes* (shortage → need redistribution **to** them)
- consistently *gain bikes* (overflow → bikes can be redistributed **from** them)

---

### Why this matters

The goal is to reduce bike shortages at high-demand stations by relocating bikes **once per day** from nearby overflow stations, using the **existing relocation fleet**.

This allows CitiBike to:

- ensure bike availability where riders need them the most  
- reduce operational costs by prioritizing only imbalance hotspots  
- improve customer satisfaction during peak commuter hours

# Top 20% of stations that handle ≈ 80% of all trips (departures & arrivals)

### Top Start Stations (rentals)

```python
# Count how many trips start at each station
starts = df['start_station_name'].value_counts()

# Calculate cumulative share of all starts
cum_share_starts = starts.cumsum() / starts.sum()

# Select top start stations that together make up 80% of rentals
top_80_start = starts[cum_share_starts <= 0.8]

print(f"TOP START STATIONS: {len(top_80_start)} stations ({len(top_80_start)/len(starts):.1%} of all) cover 80% of rentals.")
top_80_start.head()
```

### Top End Stations (returns)

```python
# 1️⃣ Exclude invalid or missing end stations first
df_valid = df[df['end_station_name'].notna() & (df['end_station_name'] != 'nan')]

# 2️⃣ Count how many trips end at each station (on clean data)
ends = df_valid['end_station_name'].value_counts()

# 3️⃣ Calculate cumulative share of all ends
cum_share_ends = ends.cumsum() / ends.sum()

# 4️⃣ Select top end stations that together make up 80% of returns
top_80_end = ends[cum_share_ends <= 0.8]

# 5️⃣ Print summary
print(f"TOP END STATIONS: {len(top_80_end)} stations ({len(top_80_end)/len(ends):.1%} of all) cover 80% of returns.")
top_80_end.head()
```

```python
# Convert to sets for easy comparison
set_start = set(top_80_start.index)
set_end = set(top_80_end.index)

# Stations that are in both top 80% (busy overall)
common = set_start & set_end

# Stations mostly popular for starting trips (origins)
only_start = set_start - set_end

# Stations mostly popular for ending trips (destinations)
only_end = set_end - set_start

print(f"Common stations (both top 80%): {len(common)}")
print(f"Top start-only stations: {len(only_start)}")
print(f"Top end-only stations: {len(only_end)}")

# Optional: preview examples
print("\nSample start-only stations:", list(only_start)[:10])
print("\nSample end-only stations:", list(only_end)[:10])
```

The CitiBike system is well-balanced overall, with only a few directional stations showing location-specific behavior.

621 are both strong origins and destinations — meaning bikes are circulating efficiently across most of the city.

Only 10 stations (≈1.6%) show one-way behavior:

5 start-only → people mainly take bikes there but rarely return them.

5 end-only → people mostly arrive there but don’t start trips from them.

Localized rebalancing:
The 10 unbalanced stations pinpoint specific residential and destination clusters that could benefit from:

Targeted bike redistribution in peak hours.

Dock expansion or relocation near those areas.

Urban pattern insight:
Start-only = residential outskirts, End-only = central business/tourist areas, confirming the commuter flow from Uptown/Brooklyn → Midtown/Downtown Manhattan.

# Top 631 stations Net flow

```python
# Count starts and ends separately
starts = df_top_stations['start_station_name'].value_counts().rename('starts')
ends = df_top_stations['end_station_name'].value_counts().rename('ends')

# Combine them into one DataFrame (outer join keeps all stations)
compare = starts.to_frame().join(ends.to_frame(), how='outer').fillna(0)

# Calculate the net difference: positive = more rentals, negative = more returns
compare['net_flow'] = compare['starts'] - compare['ends']

# Sort by imbalance (largest difference first)
compare_sorted = compare.sort_values(by='net_flow', ascending=False)

# Inspect both extremes
print("Stations with many rentals but few returns:")
print(compare_sorted.head(15))

print("\n Stations with many returns but few rentals:")
print(compare_sorted.tail(15))
```

Station Flow Interpretation:

net_flow > 0: More bikes leave than arrive — possible bike shortage. Typical for commuter origin areas such as residential neighborhoods.

net_flow < 0: More bikes arrive than leave — possible bike overflow. Typical for commuter destination areas such as office or tourist zones.

# Identify persistently unbalanced stations

```python
# Group by station and date
daily_flow = (
    df_top_stations.groupby(['start_station_name', 'date'], as_index=False)
    .size()
    .rename(columns={'start_station_name': 'station_name', 'size': 'starts'})
)

# Count ends separately
daily_ends = (
    df_top_stations.groupby(['end_station_name', 'date'], as_index=False)
    .size()
    .rename(columns={'end_station_name': 'station_name', 'size': 'ends'})
)

# Merge starts + ends by station and date
df_daily = pd.merge(daily_flow, daily_ends, on=['station_name', 'date'], how='outer').fillna(0)

# Calculate daily net flow
df_daily['net_flow'] = df_daily['starts'] - df_daily['ends']

# Aggregate by station across all days
station_balance = (
    df_daily.groupby('station_name', as_index=False)
    .agg(mean_net_flow=('net_flow', 'mean'),
         std_net_flow=('net_flow', 'std'))
)

# Flag stations with persistent imbalance 
threshold = 5  # bikes/day 
station_balance['imbalance_flag'] = station_balance['mean_net_flow'].abs() > threshold

# Extract top donor (shortage) and receiver (overflow) stations
donor_stations = station_balance[station_balance['mean_net_flow'] > threshold].sort_values('mean_net_flow', ascending=False)
receiver_stations = station_balance[station_balance['mean_net_flow'] < -threshold].sort_values('mean_net_flow', ascending=True)

# Preview
print("Stations needing deliveries (bike shortage):")
print(donor_stations.head(10))

print("\n Stations needing removals (bike overflow):")
print(receiver_stations.head(10))
```

```python
station_balance['imbalance_flag'].value_counts()
```

583 stations (≈ 92%) have imbalance_flag = False →
their average daily net flow (mean_net_flow) is within ±5 bikes/day.
These are self-balancing stations — users naturally move bikes back and forth.



48 stations (≈ 8%) have imbalance_flag = True →
their average daily imbalance exceeds ±5 bikes/day.
These are problem stations where natural flow does not recover —
they require truck-based redistribution.

# Exploring nan values in end_station_name

During data cleaning, several rides were found to have missing or invalid **end station names** (`nan`).  
These represent trips where bikes were checked out but not properly recorded as returned to any dock.

The goal of this step is to understand whether these missing endpoints could be linked to **dock saturation issues** in the city.
Stations with a high number of `nan` returns might correspond to areas where riders could not find an available docking space due to full capacity.  

1. **Filter trips** where `end_station_name` is `"nan"`.  
2. **Count start locations** of these trips to find stations most often associated with unreturned bikes.  
3. **Compare spatially** these start stations with nearby **receiver stations** (those showing persistent negative `net_flow`) — typically within a **2–4 km** radius.  
4. If proximity is confirmed, this supports the hypothesis that the missing returns result from **dock overload**, not data loss or user error.

```python
df[df['end_station_name'] == 'nan'].value_counts()
```

```python
df_nan = df[df['end_station_name'].astype(str).str.lower() == 'nan']
```

```python
df_nan['rideable_type'].value_counts()
```

```python
df_nan['start_station_name'].value_counts().sort_values(ascending=False).head(20)
```

Comparison 

Stations with persistent negative net flow (e.g., Old Slip & South St, Washington Square E) are located 2–4 km south of the busiest departure hubs in Midtown. This suggests that riders traveling from residential or commercial Midtown areas often reach already-full docking stations in Downtown and East Village.

Recommendation:
Implement dynamic rebalancing between 7–9 a.m. and 5–7 p.m., moving bikes from overflow zones in the south (receivers) back to Midtown donor stations (origins). Consider predictive alerts for dock capacity in those 2–4 km corridors.

Hourly Patterns and Spatial Context

The analysis of trips without a recorded end station reveals that most of these rides begin during the evening rush hours (4 PM – 7 PM). The single largest spike occurs at 5 PM (180 K trips), followed by 6 PM and 4 PM, indicating that users most often start such rides while returning home or heading to leisure destinations after work.

Geographically, these incomplete trips predominantly originate from Midtown Manhattan and Central Park South, including key stations such as Broadway & W 41 St, 6 Ave & W 34 St, and Grand Army Plaza & Central Park S. These are dense commercial and commuter zones surrounded by offices, hotels, and major transit hubs.

This pattern strongly suggests evening dock saturation: when users attempt to return bicycles to high-traffic central stations, many docks are already full, forcing riders to end trips manually without docking — resulting in missing end-station data.

#### Implementing **dynamic bike rebalancing** between **7–9 a.m.** and **5–7 p.m.**, moving bikes from overloaded southern docks (receivers) back to Midtown donor stations (origins), could reduce both **dock unavailability** and the **number of nan end trips**.
