```python
cols_needed = [
    'start_station_name', 'end_station_name',
    'start_lat', 'start_lng', 'end_lat', 'end_lng',
    'member_casual', 'rideable_type',
    'hour', 'weekday', 'month', 'day_type',
    'value', 'ride_duration'
]

df_waterfront = df[cols_needed]
```

```python
# Create a new column with the ride start hour
df_time['hour'] = df_time['started_at'].dt.hour
```

```python
df_time.head()
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.countplot(
    data=df_time,
    x='hour',
    hue='rideable_type',
    palette = 'crest_r'
)

plt.title('Hourly Distribution of Bike Rides — by Bike Type')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.legend(title='Bike Type')
plt.tight_layout()
plt.show()
```

```python
# Aggregate ride counts by hour and user type
hourly_counts = df_time.groupby(['member_casual', 'hour'])['ride_id'].count().reset_index()

# Create a FacetGrid — make it wider with 'aspect'
g = sns.FacetGrid(hourly_counts, col='member_casual', height=4, aspect=1.8, sharey=False)
g.map_dataframe(sns.barplot, x='hour', y='ride_id')

# Customize titles and labels
g.set_titles("{col_name}")
g.set_axis_labels("Hour of Day", "Number of Rides")

plt.suptitle("Hourly Ride Patterns — Members vs Casual Users", y=1.05, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 1️⃣ Aggregate rides by user type, hour, and bike type
hourly_counts = (
    df_time
    .groupby(['member_casual', 'hour', 'rideable_type'])['ride_id']
    .count()
    .reset_index()
)

# 2️⃣ Plot using Seaborn FacetGrid
palette = sns.color_palette("Set2", n_colors=3)  # consistent color set for bike types

g = sns.FacetGrid(
    hourly_counts,
    col='member_casual',
    height=4,
    aspect=1.8,
    sharey=False
)

# 3️⃣ Draw stacked-style bars (hue by rideable_type)
g.map_dataframe(
    sns.barplot,
    x='hour',
    y='ride_id',
    hue='rideable_type',
    palette=palette
)

# 4️⃣ Customize titles, labels, and legend
g.set_titles("{col_name}")
g.set_axis_labels("Hour of Day", "Number of Rides")
g.add_legend(title='Bike Type')

plt.suptitle("Hourly Ride Patterns — Members vs Casual Users (by Bike Type)", 
             y=1.05, fontsize=14)

plt.tight_layout()
plt.show()
```

```python
hourly_counts.to_csv("hours_member_casual.csv")
```

```python
import plotly.express as px

df_heat = df_time.groupby(["month", "hour"]).size().reset_index(name="rides")

fig = px.density_heatmap(
    df_heat,
    x="hour",
    y="month",
    z="rides",
    color_continuous_scale="Blues",
    title="CitiBike Rides by Hour and Month"
)
fig.update_layout(xaxis=dict(dtick=1))
fig.show()
```
